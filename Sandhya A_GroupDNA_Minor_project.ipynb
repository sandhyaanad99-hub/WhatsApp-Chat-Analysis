# Feature 1 - Chat Parser

```python
from datetime import datetime

def parse_chat(lines):
    messages = []
    media_counts = {}
    skipped_system = 0
    skipped_deleted = 0

    for line in lines:
        line = line.strip()

        # Skip empty lines
        if line == "":
            continue

        # Skip invalid lines
        if " - " not in line:
            continue

        timestamp, rest = line.split(" - ", 1)

        # System message (no sender)
        if ": " not in rest:
            skipped_system += 1
            continue

        sender, message = rest.split(": ", 1)

        # Count media messages
        if message == "<Media omitted>":
            media_counts[sender] = media_counts.get(sender, 0) + 1

        # Count deleted messages
        if message == "This message was deleted":
            skipped_deleted += 1

        # Store ALL non-system messages
        messages.append({
            "timestamp": timestamp,
            "sender": sender,
            "text": message
        })

    return messages, media_counts, skipped_system, skipped_deleted


# Run parser
messages, media_counts, skipped_system, skipped_deleted = parse_chat(lines)

print("=" * 60)
print("CHAT PARSER")
print("=" * 60)
print(f"Parsed Messages : {len(messages)}")
print(f"System Messages : {skipped_system}")
print(f"Media Messages  : {sum(media_counts.values())}")
print(f"Deleted Messages: {skipped_deleted}")
print(f"Media per Person: {media_counts}")
```

### Feature 2: Group Overview

```python
from collections import Counter
from datetime import datetime

def group_overview(messages, media_counts):
    print("=" * 60)
    print("                     GROUP OVERVIEW")
    print("=" * 60)

    # Total messages
    total_messages = len(messages)

    # Members and message count
    message_count = Counter(msg["sender"] for msg in messages)
    members = len(message_count)

    # Group period
    dates = []
    for msg in messages:
        try:
            dates.append(datetime.strptime(msg["timestamp"], "%d/%m/%y, %I:%M %p"))
        except:
            pass

    if dates:
        start_date = min(dates).strftime("%d/%m/%y")
        end_date = max(dates).strftime("%d/%m/%y")
    else:
        start_date = "-"
        end_date = "-"

    print(f"\nGroup      : Loca ЁЯдк gang ЁЯШ╡тАНЁЯТлЁЯла")
    print(f"Period     : {start_date} to {end_date}")
    print(f"Messages   : {total_messages}")
    print(f"Members    : {members}")

    print("\nMESSAGES PER PERSON")
    print("-" * 60)

    sorted_people = sorted(message_count.items(),
                           key=lambda x: x[1],
                           reverse=True)

    max_msgs = sorted_people[0][1]

    for person, count in sorted_people:
        percent = (count / total_messages) * 100
        bar = "тЦИ" * int((count / max_msgs) * 20)
        print(f"{person:<30} {bar:<20} {count:>4} ({percent:.1f}%)")

    print("\nMEDIA SHARED PER PERSON")
    print("-" * 60)

    if media_counts:
        media_sorted = sorted(media_counts.items(),
                              key=lambda x: x[1],
                              reverse=True)

        for person, count in media_sorted:
            print(f"{person:<30} : {count} media messages")
    else:
        print("No media messages found.")

    return message_count


# Run Feature 2
message_count = group_overview(messages, media_counts)
```

# Feature 3- Activity Analysis

```python
from collections import Counter
from datetime import datetime

def activity_analysis(messages):
    print("=" * 60)
    print("                 ACTIVITY ANALYSIS")
    print("=" * 60)

    # Messages by day of week
    day_counter = Counter()

    # Messages by month
    month_counter = Counter()

    # Messages by hour
    hour_counter = Counter()

    for msg in messages:
        try:
            dt = datetime.strptime(msg["timestamp"], "%d/%m/%y, %I:%M %p")

            day_counter[dt.strftime("%A")] += 1
            month_counter[dt.strftime("%B")] += 1
            hour_counter[dt.hour] += 1

        except:
            continue

    print("\nMOST ACTIVE DAYS")
    print("-" * 60)

    for day, count in sorted(day_counter.items(),
                             key=lambda x: x[1],
                             reverse=True):
        print(f"{day:<12}: {count} messages")

    print("\nMOST ACTIVE MONTHS")
    print("-" * 60)

    for month, count in sorted(month_counter.items(),
                               key=lambda x: x[1],
                               reverse=True):
        print(f"{month:<12}: {count} messages")

    print("\nMESSAGES BY HOUR")
    print("-" * 60)

    for hour in range(24):
        count = hour_counter.get(hour, 0)
        bar = "тЦИ" * (count // 20)
        print(f"{hour:02d}:00 - {count:4} {bar}")

    busiest_day = max(day_counter, key=day_counter.get)
    busiest_month = max(month_counter, key=month_counter.get)
    busiest_hour = max(hour_counter, key=hour_counter.get)

    print("\nSUMMARY")
    print("-" * 60)
    print(f"Busiest Day   : {busiest_day}")
    print(f"Busiest Month : {busiest_month}")
    print(f"Busiest Hour  : {busiest_hour:02d}:00")

    return day_counter, month_counter, hour_counter


# Run Feature 3
day_counter, month_counter, hour_counter = activity_analysis(messages)
```

# Feature 4 :

```python
import numpy as np
from datetime import datetime

def activity_heatmap(messages, message_count):
    print("=" * 70)
    print("                FEATURE 4 : ACTIVITY HEATMAP")
    print("=" * 70)

    # Participant names
    people = list(message_count.keys())

    # Create n x 24 NumPy matrix
    heatmap = np.zeros((len(people), 24), dtype=int)

    # Map person -> row index
    person_index = {person: i for i, person in enumerate(people)}

    # Fill the matrix
    for msg in messages:
        try:
            dt = datetime.strptime(msg["timestamp"], "%d/%m/%y, %I:%M %p")
            row = person_index[msg["sender"]]
            hour = dt.hour
            heatmap[row][hour] += 1
        except:
            continue

    print("\nNUMPY MATRIX SHAPE:", heatmap.shape)

    print("\nTEXT HEATMAP")
    print("-" * 70)

    print(f'{"Person":<25}', end="")
    for h in range(24):
        print(f"{h:02}", end=" ")
    print()

    for i, person in enumerate(people):
        print(f"{person[:24]:<25}", end="")

        max_val = np.max(heatmap[i])

        for value in heatmap[i]:
            if max_val == 0:
                block = ". "
            else:
                ratio = value / max_val

                if ratio == 0:
                    block = ". "
                elif ratio <= 0.25:
                    block = ". "
                elif ratio <= 0.50:
                    block = "тЦС "
                elif ratio <= 0.75:
                    block = "тЦТ "
                else:
                    block = "тЦИ "

            print(block, end="")

        print()

    print("\nNIGHT OWL CHECK")
    print("-" * 70)

    for i, person in enumerate(people):

        total = np.sum(heatmap[i])

        night = np.sum(
            heatmap[i, [23, 0, 1, 2, 3, 4]]
        )

        percent = (night / total) * 100 if total > 0 else 0

        if percent >= 75:
            print(f"{person:<25} ЁЯМЩ NIGHT OWL ({percent:.1f}%)")
        else:
            print(f"{person:<25} {percent:.1f}% night activity")

    return heatmap


# Run Feature 4
heatmap = activity_heatmap(messages, message_count)
```

# Feature 5-Top Words

```python
import string

def word_frequency(messages):
    print("=" * 60)
    print("              FEATURE 5 : WORD FREQUENCY")
    print("=" * 60)

    # Common stop words
    stop_words = {
        "i","is","the","a","an","and","or","to","of","in","on","for",
        "it","its","im","i'm","you","your","yours","me","my","we","our",
        "ours","he","she","they","them","this","that","these","those",
        "are","am","was","were","be","been","being","do","does","did",
        "have","has","had","will","would","can","could","should","shall",
        "with","at","by","from","as","if","but","so","not","no","yes",
        "ok","okay","ya","yeah","hai","haan","hi","hello"
    }

    word_count = {}

    # Count words
    for msg in messages:
        words = msg["text"].split()

        for word in words:
            word = word.lower()

            # Remove punctuation
            word = word.strip(string.punctuation)

            # Skip empty and stop words
            if word == "" or word in stop_words:
                continue

            word_count[word] = word_count.get(word, 0) + 1

    # Sort words
    top_words = sorted(word_count.items(),
                       key=lambda x: x[1],
                       reverse=True)[:10]

    print("\nTOP 10 MOST USED WORDS")
    print("-" * 60)

    if not top_words:
        print("No words found.")
        return word_count

    max_count = top_words[0][1]

    for word, count in top_words:
        bar = "тЦИ" * int((count / max_count) * 25)
        print(f"{word:<15} {bar:<25} {count}")

    return word_count


# Run Feature 5
word_count = word_frequency(messages)
```

# Feature 6- Response Speed and Silent Streak

```python
from datetime import datetime, timedelta

def response_and_streaks(messages, message_count):

    print("=" * 60)
    print("          RESPONSE PATTERNS")
    print("=" * 60)

    # ---------- Average Response Time ----------
    response_times = {person: [] for person in message_count}

    for i in range(1, len(messages)):

        prev = messages[i-1]
        curr = messages[i]

        if prev["sender"] != curr["sender"]:

            try:
                t1 = datetime.strptime(prev["timestamp"], "%d/%m/%y, %I:%M %p")
                t2 = datetime.strptime(curr["timestamp"], "%d/%m/%y, %I:%M %p")

                gap = (t2 - t1).total_seconds() / 60

                if gap > 0:
                    response_times[curr["sender"]].append(gap)

            except:
                pass

    avg_response = {}

    for person in response_times:

        if len(response_times[person]) > 0:
            avg = sum(response_times[person]) / len(response_times[person])
        else:
            avg = 0

        avg_response[person] = avg

    sorted_response = sorted(avg_response.items(),
                             key=lambda x: x[1])

    print("\nFASTEST / SLOWEST REPLIER")
    print("-"*60)

    if len(sorted_response) > 0:
        print(f"Fastest Replier : {sorted_response[0][0]} ({sorted_response[0][1]:.2f} mins)")
        print(f"Slowest Replier : {sorted_response[-1][0]} ({sorted_response[-1][1]:.2f} mins)")

    print("\nALL RESPONSE TIMES")
    print("-"*60)

    for person, avg in sorted_response:
        print(f"{person:<30} {avg:.2f} mins")

    # ---------- Longest Silent Streak ----------

    first_day = datetime.strptime(messages[0]["timestamp"], "%d/%m/%y, %I:%M %p").date()
    last_day = datetime.strptime(messages[-1]["timestamp"], "%d/%m/%y, %I:%M %p").date()

    all_days = []

    current = first_day

    while current <= last_day:
        all_days.append(current)
        current += timedelta(days=1)

    streaks = {}

    print("\nLONGEST SILENT STREAKS")
    print("-"*60)

    for person in message_count:

        active_days = set()

        for msg in messages:

            if msg["sender"] == person:

                day = datetime.strptime(
                    msg["timestamp"],
                    "%d/%m/%y, %I:%M %p"
                ).date()

                active_days.add(day)

        longest = 0
        current_streak = 0

        for day in all_days:

            if day not in active_days:
                current_streak += 1

                if current_streak > longest:
                    longest = current_streak

            else:
                current_streak = 0

        streaks[person] = longest

    sorted_streaks = sorted(streaks.items(),
                            key=lambda x: x[1],
                            reverse=True)

    for person, streak in sorted_streaks:
        print(f"{person:<30} {streak} days")

    return avg_response, streaks, all_days


# Run Feature 6
avg_response, streaks, all_days = response_and_streaks(messages, message_count)
```

# Feature 7- - Personality Archetype Detection

```python
from datetime import datetime

def detect_archetypes(messages, message_count, heatmap, streaks):

    print("=" * 70)
    print("              FEATURE 7 : PERSONALITY ARCHETYPES")
    print("=" * 70)

    caring_words = [
        "take care","eat","sleep","drink water","safe",
        "care","please","reminder","okay","ok"
    ]

    funny_words = [
        "lol","lmao","haha","hehe","ЁЯШВ","ЁЯдг","ЁЯШн","ЁЯШЖ"
    ]

    deep_words = [
        "life","future","why","meaning","dream",
        "think","thinking","alone","time","love"
    ]

    people = list(message_count.keys())

    for person in people:

        person_msgs = [m for m in messages if m["sender"] == person]

        total = len(person_msgs)

        if total == 0:
            continue

        # ---------------- SPAMMER ----------------

        burst_lengths = []
        burst = 0

        for msg in messages:

            if msg["sender"] == person:
                burst += 1
            else:
                if burst > 0:
                    burst_lengths.append(burst)
                burst = 0

        if burst > 0:
            burst_lengths.append(burst)

        avg_burst = sum(burst_lengths) / len(burst_lengths) if burst_lengths else 1

        # ---------------- NIGHT OWL ----------------

        row = people.index(person)

        total_msgs = heatmap[row].sum()

        night_msgs = (
            heatmap[row][23] +
            heatmap[row][0] +
            heatmap[row][1] +
            heatmap[row][2] +
            heatmap[row][3] +
            heatmap[row][4]
        )

        night_pct = night_msgs / total_msgs if total_msgs else 0

        # ---------------- GROUP MOM ----------------

        caring = 0

        for m in person_msgs:

            text = m["text"].lower()

            if any(word in text for word in caring_words):
                caring += 1

        caring_pct = caring / total

        # ---------------- STORYTELLER ----------------

        total_words = sum(len(m["text"].split()) for m in person_msgs)

        avg_words = total_words / total

        # ---------------- DRAMA QUEEN ----------------

        caps = sum(
            1 for m in person_msgs
            if len(m["text"]) > 3 and m["text"].isupper()
        )

        caps_pct = caps / total

        # ---------------- GHOST ----------------

        ghost_score = streaks.get(person, 0)

        # ---------------- COMEDIAN ----------------

        funny = 0

        for m in person_msgs:

            text = m["text"].lower()

            if any(word in text for word in funny_words):
                funny += 1

        funny_pct = funny / total

        # ---------------- QUESTION MASTER ----------------

        question = sum(
            1 for m in person_msgs
            if m["text"].strip().endswith("?")
        )

        question_pct = question / total

        # ---------------- BONUS ----------------

        philosopher = 0

        for m in person_msgs:

            hour = datetime.strptime(
                m["timestamp"],
                "%d/%m/%y, %I:%M %p"
            ).hour

            text = m["text"].lower()

            if hour <= 4 and any(word in text for word in deep_words):
                philosopher += 1

        philosopher_pct = philosopher / total

        # ---------------- SCORE ----------------

        scores = {
            "THE SPAMMER": avg_burst / 10,
            "THE NIGHT OWL": night_pct,
            "THE GROUP MOM": caring_pct,
            "THE STORYTELLER": avg_words / 40,
            "THE DRAMA QUEEN": caps_pct,
            "THE GHOST": ghost_score / 30,
            "THE COMEDIAN": funny_pct,
            "THE QUESTION MASTER": question_pct,

            # Bonus Archetype
            "THE LATE NIGHT PHILOSOPHER": philosopher_pct
        }

        archetype = max(scores, key=scores.get)

        print(f"{person:<30} тЖТ {archetype}")

    print("=" * 70)
detect_archetypes(
    messages,
    message_count,
    heatmap,
    streaks
)
```

# Feature 8 - The Final Report

```python
def generate_report(file_name, start_date, end_date, total_days, total_messages, total_users, avg_per_day, top_users, hourly_dist):
    """
    Generates a terminal report where data lists are shown first,
    followed collectively by their corresponding visual chart sections.
    """
    w = 70  # Total content width inside borders

    # Border Characters
    top_b    = "тФМ" + "тФА" * w + "тФР"
    mid_b    = "тФЬ" + "тФА" * w + "тФд"
    bot_b    = "тФФ" + "тФА" * w + "тФШ"
    sep_line = "тФВ " + "тФА" * (w - 2) + " тФВ"

    def row(text=""):
        return f"тФВ {text:<{w-2}} тФВ"

    # --- REPORT START ---
    print(top_b)
    print(row(f"{'WHATSAPP CHAT ANALYSIS REPORT':^68}"))
    print(mid_b)

    # Metadata Section
    print(row(f"Source File: {file_name}"))
    print(row(f"Time Range:  {start_date} to {end_date} ({total_days} days)"))
    print(mid_b)

    # High-Level Stats Section
    print(row(f"{'HIGH-LEVEL STATS':^68}"))
    print(sep_line)
    print(row(f"Total Messages: {total_messages:<15,} Unique Users: {total_users}"))
    print(row(f"Avg Msg / Day:  {avg_per_day:<15.1f}"))
    print(mid_b)

    # --- TOP USERS SECTION ---
    print(row(f"{'TOP 5 MOST ACTIVE USERS':^68}"))
    print(sep_line)

    # 1. First print the text list data
    for rank, (user, count) in enumerate(top_users[:5], 1):
        share = (count / total_messages) * 100
        print(row(f" {rank}. {user:<22} тФВ {count:>6,} msgs тФВ {share:>5.1f}%"))

    print(row("")) # Clean divider space

    # 2. Then print the complete visual chart below it
    max_user_msgs = top_users[0][1] if top_users else 1
    for rank, (user, count) in enumerate(top_users[:5], 1):
        bar_len = int((count / max_user_msgs) * 45) if max_user_msgs > 0 else 0
        bar = "тЦИ" * max(1, bar_len) if count > 0 else ""
        print(row(f" [{rank}] {bar}"))

    print(mid_b)

    # --- HOURLY DISTRIBUTION SECTION ---
    print(row(f"{'HOURLY ACTIVITY DISTRIBUTION':^68}"))
    print(sep_line)

    time_blocks = [
        ("Night (00-06)", list(range(0, 6))),
        ("Morning (06-12)", list(range(6, 12))),
        ("Afternoon (12-18)", list(range(12, 18))),
        ("Evening (18-24)", list(range(18, 24)))
    ]

    block_totals = [sum(hourly_dist.get(h, 0) for h in hours) for _, hours in time_blocks]
    max_block_sum = max(block_totals) if block_totals else 1

    # 1. First print the text list data
    for (block_name, _), block_sum in zip(time_blocks, block_totals):
        block_share = (block_sum / total_messages) * 100
        print(row(f" тАв {block_name:<18} тФВ {block_sum:>6,} msgs тФВ {block_share:>5.1f}%"))

    print(row("")) # Clean divider space

    # 2. Then print the complete visual density chart below it
    for (block_name, _), block_sum in zip(time_blocks, block_totals):
        bar_len = int((block_sum / max_block_sum) * 45) if max_block_sum > 0 else 0
        bar = "тЦС" * max(1, bar_len) if block_sum > 0 else ""
        # Using a shortened label prefix for clean alignment
        prefix = block_name.split()[0]
        print(row(f" {prefix:<10} тФВ {bar}"))

    print(bot_b)

# --- SIMULATION RUN ---
sample_hourly_data = {h: 150 if h < 6 else (450 if h < 12 else (1100 if h < 18 else 1800)) for h in range(24)}

generate_report(
    file_name="WhatsApp Chat with Loca ЁЯдк gang ЁЯШ╡тАНЁЯТлЁЯла.txt",
    start_date="2024-01-15",
    end_date="2026-06-25",
    total_days=892,
    total_messages=14241,
    total_users=6,
    avg_per_day=16.0,
    top_users=[
        ("Loca Leader ЁЯдк", 5420),
        ("User Two ЁЯШ╡тАНЁЯТл", 3850),
        ("User Three ЁЯла", 2410),
        ("Friend Four", 1560),
        ("Group Member 5", 1001)
    ],
    hourly_dist=sample_hourly_data
)
```
